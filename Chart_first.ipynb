```python
import matplotlib.pyplot as plt

%matplotlib inline
from scipy.interpolate import make_interp_spline
import numpy as np
y = [660,690,665,680,650,680,700] 
# x = ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday'] 
x = ['Aug 2022','Sept 2022','Oct 2022','Nov 2022', 'Dec 2022','Jan 2023','Feb 2023']
plt.figure(facecolor='lightgreen')
plt.plot(x,y,linestyle='solid',linewidth=2, markersize=5)
# ax = plt.axes()
# ax.set_facecolor("lightgreen")
plt.xlabel('Days')
plt.ylabel('No. of views ')
plt.title('Page Views')
```

```python
companies = ('Amazon', 'Flipkart', 'Meesho')
reviews = {
    'Positive': np.array([80, 50, 40]),
    'Neutral': np.array([52, 40, 58]),
    'Negative': np.array([23, 40, 58]),
}
width = 0.3 # the width of the bars: can also be len(x) sequence

# plt.figure(facecolor='lightgreen')
fig, ax = plt.subplots()
bottom = np.zeros(3)

for review, reviews in reviews.items():
    p = ax.bar(companies, reviews, width, label=review, bottom=bottom)
    bottom += reviews

    ax.bar_label(p, label_type='center')

ax.set_title("Review's split Percentage")
ax.legend()
fig.patch.set_facecolor('lightgreen')

plt.show()
```
